### How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.

```python
race_count = df['race'].value_counts()
race_count
```

```python
# What is the average age of men?
average_age_men = df[df['sex'] == 'Male']['age'].mean()
average_age_men
```

### What is the percentage of people who have a Bachelor's degree?

```python
percentage_bachelors = df[df['education'] == 'Bachelors'].shape[0] * 100 / df.shape[0]
print(percentage_bachelors,f'%')
```

### What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?
### What percentage of people without advanced education make more than 50K?

```python
# with and without `Bachelors`, `Masters`, or `Doctorate`
higher_education = df[df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
lower_education = df[~df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])] 
```

```python
lower_education
```

```python
higher_education
```

###  Percentage with salary >50K

```python
higher_education_rich = higher_education[higher_education['salary'] == '>50K']['salary'].count() / higher_education.shape[0] * 100
lower_education_rich = lower_education[lower_education['salary'] == '>50K']['salary'].count() / lower_education.shape[0] * 100
print (higher_education_rich, f'%,',lower_education_rich,f'%') 
```

###  What is the minimum number of hours a person works per week (hours-per-week feature)?

```python
min_work_hours = df['hours-per-week'].min()
min_work_hours
```

### What percentage of the people who work the minimum number of hours per week have a salary of >50K?

```python
num_min_workers = df[df['hours-per-week'] == 1]['hours-per-week'].sum()
num_min_workers
```

```python
rich_percentage = df[(df['hours-per-week'] == 1) & (df['salary'] == '>50K')].shape[0] *100 / num_min_workers
print (rich_percentage, f'%') 
```

```python
# Changing '?' for 'other' (optional)
df['native-country'] = df['native-country'].apply(lambda x: x.strip())
country_stats = df.groupby('native-country')['native-country'].agg('count').sort_values(ascending=False)
country_stats
```

```python
df2 =  {'?': 'Other'}
df['native-country'] = [df2.get(n,n) for n in df['native-country']]
df['native-country'].unique()
```

### What country has the highest percentage of people that earn >50K?

```python
all = df[['salary','native-country']]
entire = all.groupby(['native-country']).count()
plus50k = all[all['salary']=='>50k']
plus50k = plus50k.groupby(['native-country']).count()
plus50k = (plus50k*100)/entire 
plus50k.sort_values(by=['salary'], inplace=True, ascending = False)
final = pd.Series(plus50k['salary'])
highest_earning_country = final.index[0]
highest_earning_country_percentage = (final.values[0],1) 
highest_earning_country,highest_earning_country_percentage
```
